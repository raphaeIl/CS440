## Evaluating the Performance of the Bots

To make things consistent, I pre-generated 10 different 50x50 ships, each bot will be tested on those same ships with different paramters.

For each Bot, I will run the simulation 10 times for each paramter (for now) and average the Actions taken and evaluate the performance on those

I will be using a range of different k and a values below:

```python
k_values = [1, 2, 3, 4, 5]
a_values = [0.1, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5]
```

### Bot 1 vs Bot 2, as a function of k

```python
import matplotlib.pyplot as plt

bot1_data = [1078.4, 631.3, 559.7, 389, 357.9]
bot2_data = [1334.8, 846.7, 374.8, 299.4, 212.5]


plt.figure(figsize=(10, 6))
plt.plot(k_values, bot1_data, marker='o', label='Bot 1')
plt.plot(k_values, bot2_data, marker='o', label='Bot 2')

plt.xticks(k_values)

plt.title('Average Actions taken of Bots vs. k values')
plt.xlabel('k value')
plt.ylabel('Average Action')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

plt.show()
```

### Bot 3 vs Bot 4, as a function of a

```python
bot3_data = [339.3, 312.9, 122.3, 181.9, 344.5, 299.3, 380.5, 290.0, 839.2]
bot4_data = [413.4, 157.1, 223.1, 375.9, 429.1, 638.7, 1537.7, 800.4, 1348.5]


plt.figure(figsize=(10, 6))
plt.plot(a_values, bot3_data, marker='o', label='Bot 3')
plt.plot(a_values, bot4_data, marker='o', label='Bot 4')

plt.xticks(a_values)

plt.title('Average Actions taken of Bots vs. a values')
plt.xlabel('a value')
plt.ylabel('Average Action')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

plt.show()
```
