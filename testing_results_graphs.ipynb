### After Experimenting a wide range of values for ship_size, q, on different bots. I will analyze all the data I got

#### q values: [0, 0.25, 0.5, 0.75, 1]
#### ship_size: 20
#### Bots: 1, 2, 3, 4

Due to time issues, I was only able to run 100 tests per bot for each q

```python
import matplotlib.pyplot as plt

# Hyperparameter values
q_values = [0, 0.25, 0.5, 0.75, 1]

bot1_data = [0.97, 0.92, 0.85, 0.81, 0.74]
bot2_data = [0.99, 0.97, 0.93, 0.84, 0.81]
bot3_data = [1.00, 0.96, 0.93, 0.91, 0.88]
bot4_data = [1.00, 0.95, 0.94, 0.92, 0.91]


plt.figure(figsize=(10, 6))
plt.plot(q_values, bot1_data, marker='o', label='Bot 1')
plt.plot(q_values, bot2_data, marker='o', label='Bot 2')
plt.plot(q_values, bot3_data, marker='o', label='Bot 3')
plt.plot(q_values, bot4_data, marker='o', label='Bot 4')

plt.xticks(q_values)

plt.title('Success Rate of Bots vs. q values (100 Experiments, ship_size=20)')
plt.xlabel('q value')
plt.ylabel('Success Rate')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.tight_layout()

# Display the plot
plt.show()
```
